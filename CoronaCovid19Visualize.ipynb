```python
import datetime
import csv
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
```

```python
CovidDataset= pd.read_csv("covid_19_dataset.csv", delimiter ="," )
```

```python
CovidDataset
```

```python
JanuaryNepalData=CovidDataset[(CovidDataset.Country == 'Nepal') & (CovidDataset.Month == 1)]
```

```python
FebruaryNepalData=CovidDataset[(CovidDataset.Country == 'Nepal') & (CovidDataset.Month == 2)]
```

```python
JanuaryNepalData
```

```python
FebruaryNepalData
```

```python
plt.title("CoronaVirus(Covid-19) outbreak in Nepal 'January 2020' \n")
plt.bar(JanuaryNepalData.Date , JanuaryNepalData.Patients / 1)
plt.xticks(JanuaryNepalData.Date)
plt.xlabel("Date -------->")
plt.ylabel("Number of Patients -------->")
plt.yticks(JanuaryNepalData.Patients)
plt.show()
plt.title("CoronaVirus(Covid-19) outbreak in Nepal 'February 2020' \n ")
plt.bar(FebruaryNepalData.Date , FebruaryNepalData.Patients  )
plt.xticks(FebruaryNepalData.Date)
plt.yticks(FebruaryNepalData.Patients)
plt.xlabel("Date -------->")
plt.ylabel("Number of Patients -------->")
plt.show()
```
